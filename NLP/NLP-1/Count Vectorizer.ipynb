````python
'''
Goal : To lemmitize the data entire corpus.

Steps
1. Create a tuple containing category and words to the document
2. clean the words of the document by removing stops words & punctuations and apply lementization.
'''
````

````python
from nltk.corpus import movie_reviews
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import string
````

````python
documents = []
for category in movie_reviews.categories():
    for files in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(files),category))
    
documents[0:5]
````

````python
from nltk.corpus import stopwords
stops = list(set(stopwords.words("english")))


punctuations = list(string.punctuation)

#updating stop words list by adding punctuations
stops = stops+punctuations
````

````python
from nltk.corpus import wordnet
def simple_pos(pos):
    if pos.startswith("V"):
        return wordnet.VERB
    elif pos.startswith('J'):
        return wordnet.ADJ
    elif pos.startswith("N"):
        return wordnet.NOUN
    elif pos.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
````

````python
output = []
lem = WordNetLemmatizer()

def clean_doc(words):
    for w in words:
        if w.lower() not in stops:
            pos        = pos_tag([w])
            clean_word = lem.lemmatize(w,pos = simple_pos(pos[0][1]))
            output.append(clean_word)
    return output
````

````python
import random
random.shuffle(documents)
````

````python
document = [(clean_doc(doc),category) for doc,category in documents]
````
