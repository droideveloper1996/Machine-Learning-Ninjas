```python
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
```

```python
olivetti_data = fetch_olivetti_faces()
```

```python
olivetti_data.keys()
```

```python
olivetti_data.images.shape  #400 images each of size 64*64
```

```python
x = olivetti_data.data
y = olivetti_data.target
```

```python
fig = plt.figure(figsize=(10,10))
for i in range(100):
    ax = fig.add_subplot(10,10,i+1)#row,col,position_number
    ax.imshow(olivetti_data.images[i])
```

```python
pca = PCA()
pca.fit_transform(x)
```

```python
pca.components_.shape  # though shape should be 4096*400 but is opposite
```

```python
pca.explained_variance_.shape  # though no of features are 64*64 = 4096 but we get 400 as features.
```

```python
k = 0
total = sum(pca.explained_variance_)
current_variance = 0
while current_variance/total <0.95:
    current_variance = current_variance + pca.explained_variance_[k]
    k+=1
    
k
```
